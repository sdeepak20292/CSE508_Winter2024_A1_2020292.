```python
def convert_to_lowercase(input_file, output_file):

      # Open the input file for reading
      with open(input_file, 'r') as file:
          # Read the contents of the file
          content = file.read()

      # Convert the content to lowercase
      lowercase_content = content.lower()

      # Open the output file for writing
      with open(output_file, 'w') as file:
          # Write the lowercase content to the output file
          file.write(lowercase_content)


```

```python
def tokenize_file(input_file, output_file):

        # Open the input file for reading
        with open(input_file, 'r') as file:
            # Read the contents of the file
            content = file.read()

        # Tokenize the content
        tokens = word_tokenize(content)

        # Open the output file for writing
        with open(output_file, 'w') as file:
            # Write the tokens to the output file
            file.write(' '.join(tokens))

```

```python
nltk.download('stopwords')
nltk.download('punkt')

def remove_stopwords(input_file, output_file):

        # Open the input file for reading
        with open(input_file, 'r') as file:
            # Read the contents of the file
            content = file.read()

        # Tokenize the content
        tokens = word_tokenize(content)

        # Remove stopwords
        stop_words = set(stopwords.words('english'))
        filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

        # Open the output file for writing
        with open(output_file, 'w') as file:
            # Write the filtered tokens to the output file
            file.write(' '.join(filtered_tokens))
```

```python
def remove_punctuation(input_file, output_file):

        # Open the input file for reading
        with open(input_file, 'r') as file:
            # Read the contents of the file
            content = file.read()

        # Remove punctuation
        cleaned_text = content.translate(str.maketrans('', '', string.punctuation))

        # Open the output file for writing
        with open(output_file, 'w') as file:
            # Write the cleaned text to the output file
            file.write(cleaned_text)

```

```python
nltk.download('punkt')

def remove_blank_space_tokens(input_file, output_file):

        # Open the input file for reading
        with open(input_file, 'r') as file:
            # Read the contents of the file
            content = file.read()

        # Tokenize the content
        tokens = word_tokenize(content)

        # Remove blank space tokens
        filtered_tokens = [word for word in tokens if word.strip()]

        # Open the output file for writing
        with open(output_file, 'w') as file:
            # Write the filtered tokens to the output file
            file.write(' '.join(filtered_tokens))

```

```python
import os
def preprocess_document(document):
    # Simple preprocessing function, you can customize it based on your needs
    document = document.lower()  # Convert to lowercase
    document = re.sub(r'[^a-z0-9\s]', '', document)  # Remove non-alphanumeric characters
    tokens = document.split()  # Tokenize the document
    return tokens

def build_inverted_index(folder_path):
    inverted_index = defaultdict(list)

    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            file_path = os.path.join(folder_path, filename)

            with open(file_path, 'r', encoding='utf-8') as file:
                document = file.read()
                tokens = preprocess_document(document)

                for position, term in enumerate(tokens):
                    inverted_index[term].append((filename, position + 1))  # position starts from 0

    return inverted_index


folder_path = '/content/drive/MyDrive/final_preprocessed'
inverted_index = build_inverted_index(folder_path)

# Print the inverted index
for term, postings in inverted_index.items():
    print(f"{term}: {postings}")
```

```python
def intersection(postings1, postings2):
    # Returns the intersection of two posting lists
    result = []
    i, j = 0, 0

    while i < len(postings1) and j < len(postings2):
        doc1, pos1 = postings1[i]
        doc2, pos2 = postings2[j]

        if doc1 == doc2:
            result.append((doc1, min(pos1, pos2)))
            i += 1
            j += 1
        elif doc1 < doc2:
            i += 1
        else:
            j += 1

    return result

def union(postings1, postings2):
    # Returns the union of two posting lists
    result = []
    i, j = 0, 0

    while i < len(postings1) or j < len(postings2):
        if i < len(postings1) and j < len(postings2):
            doc1, pos1 = postings1[i]
            doc2, pos2 = postings2[j]

            if doc1 == doc2:
                result.append((doc1, min(pos1, pos2)))
                i += 1
                j += 1
            elif doc1 < doc2:
                result.append(postings1[i])
                i += 1
            else:
                result.append(postings2[j])
                j += 1
        elif i < len(postings1):
            result.append(postings1[i])
            i += 1
        else:
            result.append(postings2[j])
            j += 1

    return result

def difference(postings1, postings2):
    result = []
    i, j = 0, 0

    while i < len(postings1) and j < len(postings2):
        doc1, pos1 = postings1[i]
        doc2, pos2 = postings2[j]

        if doc1 == doc2:
            i += 1
            j += 1
        elif doc1 < doc2:
            result.append(postings1[i])
            i += 1
        else:
            j += 1

    while i < len(postings1):
        result.append(postings1[i])
        i += 1

    return result

def perform_query(query, inverted_index):
    query = query.lower()
    query = re.sub(r'\band\b', '&', query)
    query = re.sub(r'\bor\b', '|', query)
    query = re.sub(r'\bnot\b', '~', query)
    terms = re.findall(r'\b\w+\b', query)

    result = inverted_index.get(terms[0], [])

    for i in range(1, len(terms), 2):
        operator = terms[i]
        next_term = terms[i + 1]

        if operator == '&':
            result = intersection(result, inverted_index.get(next_term, []))
        elif operator == '|':
            result = union(result, inverted_index.get(next_term, []))
        elif operator == '~':
            result = difference(result, inverted_index.get(next_term, []))

    return result
```

```python
def process_text(input_text):
    # Convert to lowercase
    lowercased_text = input_text.lower()

    # Tokenization
    tokens = word_tokenize(lowercased_text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [token for token in tokens if token not in stop_words]

    # Remove punctuations
    no_punctuations = [token for token in filtered_tokens if token not in string.punctuation]

    # Remove blank space tokens
    final_tokens = [token for token in no_punctuations if token.strip()]

    # Join tokens back into a string
    processed_text = ' '.join(final_tokens)

    return processed_text


def merge_sequence(list1, list2):
    # Join the sequence into a single string
    merged_list = []
    min_len = min(len(list1), len(list2))

    for i in range(min_len):
        merged_list.append(list1[i])
        merged_list.append(list2[i])

    # If one list is longer than the other, add the remaining elements
    merged_list.extend(list1[min_len:])
    merged_list.extend(list2[min_len:])

    q = ""
    for item in merged_list:

      q += item +" "

    return q
```

```python
N = input("Enter number of queries to perform")
for i in range(len(N)):
    input_sequence = input("Write the input sequence")
    input_operations = input("Write the operations")
    operations = input_operations.split(', ')
    seq = input_sequence.split(" ")

    # now preprocssing the input sequence.
    query = merge_sequence(seq, operations)

    result = perform_query(query, inverted_index)

    # Output format
    print(f"Query: {query}")
    print(f"Number of documents retrieved: {len(result)}")
    print("Name of Documents retrieved:")
    for doc, pos in result:
        print(f"{doc} (Position: {pos})")
    print()
```

```python
import os
def create_positional_index(folder_path):
    # Initialize positional index as a defaultdict of defaultdict
    positional_index = defaultdict(lambda: defaultdict(list))
    # Iterate through each file in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith(".txt"):
            file_path = os.path.join(folder_path, filename)
            with open(file_path, 'r', encoding='utf-8') as file:
                tokens = file.read().split()  # Assuming tokens are space-separated

            # Create positional index for each term
            for position, term in enumerate(tokens):
                positional_index[term][filename].append(position + 1)
    return positional_index

folder_path = '/content/drive/MyDrive/final_preprocessed'
positional_index = create_positional_index(folder_path)
```

```python
import nltk
nltk.download('punkt')
def process_queries(positional_index, queries):
    results = []
    for query in queries:
        query_terms = query.split()
        query_results = set()

        for term in query_terms:
            if term in positional_index:
                if not query_results:
                    query_results.update(positional_index[term])
                else:
                    query_results.intersection_update(positional_index[term])

        results.append(len(query_results))
        results.extend(sorted(query_results))

    return results
```

```python
def process_queries(positional_index, queries):
    results = []

    for query in queries:
        query_terms = query.split()
        doc_lists = []

        for term in query_terms:
            if term in positional_index:
                doc_lists.append(set(positional_index[term].keys()))

        if doc_lists:
            # Intersection of document lists for each term
            common_docs = set.intersection(*doc_lists)
            results.append((len(common_docs), common_docs))
        else:
            results.append((0, set()))

    return results
```
